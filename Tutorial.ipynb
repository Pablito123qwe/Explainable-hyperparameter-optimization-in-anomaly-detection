```python
from anomalib.data import MVTec
from anomalib.models import Patchcore
from anomalib.engine import Engine
from categories import get_categories

path = "/mnt/c/Users/pawel/Explainable-hyperparameter-optimization-in-anomaly-detection/datasets/MVTec"
train_batch_size = 2
coreset_sampling_ratio = 0.1
limit_train_batches = 3
categories = get_categories(path)

datamodule = MVTec(
    category=categories[0],
    train_batch_size=train_batch_size,
    num_workers=2,
    root=path
)
checkpoint_path = "/mnt/c/Users/pawel/Explainable-hyperparameter-optimization-in-anomaly-detection/results/Patchcore/MVTec/bottle/v15/weights/lightning/model.ckpt"
model = Patchcore.load_from_checkpoint(
    checkpoint_path,
    backbone="resnet18",
    coreset_sampling_ratio=coreset_sampling_ratio
)
engine = Engine(
    max_epochs=1,
    accelerator="cpu",
    limit_train_batches=limit_train_batches
)
test_results = engine.test(model=model, datamodule=datamodule)
print("Test results:", test_results)
```
